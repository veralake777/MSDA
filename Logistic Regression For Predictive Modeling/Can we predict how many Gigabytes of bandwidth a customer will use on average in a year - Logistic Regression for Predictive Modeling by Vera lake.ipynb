1.  Summarize the assumptions of a multiple regression model.\

    There are four assumptions of multiple regression methods:\
    a) Variables are normally distributed. This means the variables are not highly skewed or kurtotic, or include substantial outliers.\
    b) There is a linear relationship between the independent variable (IV) and the dependent variable (DV). If the relationship is not linear the results of the regression analysis will under-estimate the relationship. \
    c) Variables are measured without error reliably. \
    d) Homoscedasticity for all IVs. This means that the variance of errors is the same across all levels of the IVs. (Osborn, J. W., & Waters, E., 2002)

### D.2.  Justify a statistically based variable selection procedure and a model evaluation metric to reduce the initial model in a way that aligns with the research question.

**Variable Selection Procedure**
- Backward Feature Selection between initial and reduced OLS model

**Model Evaluation Metric**
If the P value is greater than .05 then drop the feature as this is an indicator that the independent variable is not important in predicting the dependent variable (Avg GB Usage Per Year).

```python
# Backward selection
# Resource: https://www.analyticsvidhya.com/blog/2020/10/a-comprehensive-guide-to-feature-selection-using-wrapper-methods-in-python/
def backward_elimination(data, target,significance_level = 0.05):
    features = data.columns.tolist()
    statistically_insignificant_features = []
    while(len(features)>0):
        features_with_constant = sm.add_constant(data[features])
        p_values = sm.OLS(target, features_with_constant).fit().pvalues[1:]
        max_p_value = p_values.max()
        if(max_p_value >= significance_level):
            excluded_feature = p_values.idxmax()
            features.remove(excluded_feature)
        else:
            statistically_insignificant_features.append(excluded_feature)
            break 
    return features

reduced_predictor_variables = backward_elimination(cleaned_predictor_variables, target, significance_level = 0.05)
reduced_predictor_variables = cleaned_predictor_variables[reduced_predictor_variables]
```
